```python
import pandas as pd
```

```python
df = pd.read_csv("../data/adult.data.csv", header=None)
df.columns="""age workclass fnlwgt education education-num marital-status occupation relationship race sex capital-gain capital-loss hours-per-week native-country annual-salary""".split(" ")
df.head()
```

```python
# подготовим данные
string_cols = df.select_dtypes(['object'])
df[string_cols.columns] = string_cols.apply(lambda x: x.str.strip())
income_dict = {'<=50K' : 0, '>50K' : 1}
sex_dict = {'Male' : 1, 'Female' : 0}
df['income_larger_50K'] = df['annual-salary'].map(income_dict)
df['sex'] = df['sex'].map(sex_dict)
df = df.drop('annual-salary'.split(' '), axis=1)
df.rename(columns = {'sex' : 'is_male'}, inplace=True)
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
print('Мужчин: ', sum(df[df['is_male'] == 1]['is_male']))
print('Женщин: ', df[df['is_male'] == 0]['is_male'].count())
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
print('Средний возраст женщин: ', df[df['is_male'] == 0]['age'].mean())
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
print('доля граждан Германии(в %): ', df[df['native-country'] == 'Germany'].shape[0] / df.shape[0] * 100)
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_definitions = 'Married-civ-spouse Married-spouse-absent Married-AF-spouse'.split(' ')
```

```python
married = df[df['marital-status'].isin(married_definitions)]
married_and_rich = married[married['income_larger_50K'] == 1]
married_ratio =  married_and_rich.shape[0] / married.shape[0]
```

```python
unmarried = df[~df['marital-status'].isin(married_definitions)]
unmarried_and_rich = unmarried[unmarried['income_larger_50K'] == 1]
unmarried_ratio = unmarried_and_rich.shape[0]/unmarried.shape[0]
```

```python
print('Доля зарабатывающих много (>50K) больше среди женатых мужчин:', married_ratio > unmarried_ratio)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_working_hours = df['hours-per-week'].max()
print('Максимальное число часов человек работает в неделю', max_working_hours)

men_working_max = df[df['hours-per-week'] == max_working_hours]
men_working_max_and_rich = men_working_max[men_working_max['income_larger_50K'] == 1]

print('Сколько людей работают такое количество часов:', men_working_max.shape[0])
print('Процент зарабатывающих много среди них:', men_working_max_and_rich.shape[0]/men_working_max.shape[0])
```
